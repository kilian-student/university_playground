# Visualization overview

```python
import matplotlib.pyplot as plt
```

## 1. pie charts

### Anwendungsfälle
showing proportions/share; grouping of 5 or fewer categories
### Guidelines
1. categories must conceptually sum to a single "whole"
2. number of categories should be condensed to five or fewer categories
3. proportion of data you want to highlight should be 1/4, 1/3, 1/2, 2/3 or 3/4

```python
# Data to plot
labels = 'Category 1', 'Category 2', 'Category 3', 'Category 4'
sizes = [25, 30, 20, 25]  # Percentage distribution for each category
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # Explode the first slice (Category 1)

# Plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the pie chart
plt.show()
```

## 2. bar and column charts

### Anwendungsfälle
most effective to highlight relationships/differences for discrete/nominal data
### Guidelines
1. must start at zero
2. vertical bars are good for denser data with less variation
3. horizontal bars are better for more variation and/or longer labels

```python
# Data to plot
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [30, 50, 25, 40]

# Plotting the bar chart
plt.bar(categories, values, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display the bar chart
plt.show()
```

## 3. line charts

### Anwendungsfälle
rates of change/ value differences
### Guidelines
1. data points must represent values over time
2. data lines should occupy roughly 2/3 of the vertical chart area
3. four or fewer lines should be distinctly colored/labeled

```python
# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

# Plotting the line chart
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Line Chart')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Chart')

# Adding a legend
plt.legend("palim")

# Display the line chart
plt.show()
```

## 4. scatter charts

### Anwendungsfälle
large number of data points that illustrate a clear trend/deviation
### Guidelines
1. data must be high volume enough for trends and outliers to be visible
2. relevant benchmarks, whether arising from the data or outside rules, should be visualized
3. most of the data should be a "background" color, with no more than a handful of points highlighted
4. Deckkraft anpassen oder Jitter anwenden, um die Datenverdeckung zu minimieren

```python
import numpy as np

# Generate random data for demonstration purposes
np.random.seed(42)
num_points = 50
x = np.random.rand(num_points)
y = 2 * x + 1 + 0.1 * np.random.randn(num_points)

# Plotting the scatter chart
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Chart Example')

# Adding a legend
plt.legend()

# Display the scatter plot
plt.show()
```
